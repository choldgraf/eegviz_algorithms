```python
Base = declarative_base()
Session = orm.sessionmaker()
```

```python
class Events(Base):
    __tablename__ = 'Events'
    id = Column(Integer, primary_key=True) #this may not be necessary?
    name = Column(String)
    start_time = Column(Date)
    
class Readings(Base):
    __tablename__ = "Readings"
    id = Column(Integer, primary_key=True)
    username = Column(String)
    start_time = Column(Date)
    end_time = Column(Date)
    raw_values = Column(postgresql.ARRAY(Integer))
    
class P300(Base):
    __tablename__ = "P300"
    card_name = Column(String, primary_key=True)
    avg_values = Column(postgresql.ARRAY(Integer))
    time_values = Column(postgresql.ARRAY(Date))
    auc = Column(Float)
    exp = Column(Integer)
    
```

```python
eng = sq.create_engine(SQLALCHEMY_DATABASE_URI)
Session.configure(bind=eng)
session = Session()
```

```python
qu_ev = session.query(Events)
qu_re = session.query(Readings)
```

```python
start_time = datetime.datetime(2014, 12, 8, 21, 47)
end_time = datetime.datetime(2014, 12, 8, 21, 49)
# start_time = datetime.datetime(2014, 12, 11, 02, 01, 24) # Blink times
# end_time = datetime.datetime(2014, 12, 11, 02, 01, 59)
# start_time = datetime.datetime(2014, 12, 12, 23, 57, 00)
# end_time = datetime.datetime(2014, 12, 13, 00,04, 17)
```

```python
re_events = qu_ev.filter(Events.start_time > start_time, Events.start_time < end_time)
re_data = qu_re.filter(Readings.start_time > start_time, Readings.start_time < end_time)
```

```python
# Replace timezones with UTC
results_events = []
for inst in re_events:
    stt_time = inst.start_time.replace(tzinfo=pytz.timezone('UTC'))
    results_events.append([inst.name, stt_time])
events = pd.DataFrame(results_events, columns=['event_name', 'start_time']).set_index('start_time')
events = events.sort_index()
num_mapping = {nm: i+1 for i, nm in enumerate(events.event_name.unique())}
```

```python
cols = ['time_start', 'sub_name', 'raw_value']
data_all = {}
start_times = []
for inst in re_data:
    n_pts =len(inst.raw_values)
    stt_time = inst.start_time.replace(tzinfo=pytz.timezone('UTC'))
    times = pd.date_range(start=stt_time, periods=n_pts, freq='2ms', tz='UTC')
    start_times.append(stt_time)
    for t, v in zip(times, inst.raw_values):
        try:
            data_all[inst.username][t] = v
        except KeyError:
            data_all[inst.username] = {t: v}
```

```python
fs_new = '2ms'
f_lo = 1
f_hi = 10
data = pd.DataFrame(data_all).sort_index()
data = data.replace(np.nan, 0)
data = data.apply(mne.filter.band_pass_filter, args=(Fs, f_lo, f_hi),
                  filter_length='3s')
data = data.resample(fs_new)
data.columns.name = 'name'
names = data_all.keys()
```

```python
# Overview of many events
stt_time = start_time + datetime.timedelta(seconds=100)
stp_time = stt_time + datetime.timedelta(seconds=30)

rng = slice(stt_time, stp_time)
ax = data.dropna()[rng].apply(scale).plot(figsize=(15, 5))
ax.vlines(events.index, *ax.get_ylim(), color='g', linewidth=1)
```

```python
# To explore individual events
ev_stt = events.iloc[4].name
ev_stp = ev_stt + datetime.timedelta(seconds=.75)
data.dropna().truncate(ev_stt, ev_stp).plot()
```

```python
# Only keep times w/ no null values
keep = data.isnull().sum(1).values == 0
data = data[keep]
```

```python
# If we only want to analyze a subset of trials (e.g., if multiple trials in one dataset)
cut_time = pd.datetime(2014, 12, 8, 21, 47, 45)
data = data[data.index < cut_time]
```

```python
# To pull our own trials
data_trials = []
tr_stt = -.1
tr_stp = .75
for t_num, (event_time, event_name) in enumerate(events.iterrows()):
    # Create our new trial-based time
    stt = event_time + datetime.timedelta(seconds=tr_stt)
    stp = event_time + datetime.timedelta(seconds=tr_stp)
    in_range = (data.index > stt) * (data.index < stp)
    data_trial = data[in_range]
    
    # New index for time/trials
    ix_time = np.arange(0, data_trial.shape[0]*2, 2)+tr_stt*1000
    ix_time = pd.Index(ix_time.astype(int), name='time')
    ix_trial = pd.Index([t_num]*data_trial.shape[0], name='epoch')
    ix_event = pd.Index([event_name[0]]*data_trial.shape[0], name='event')
    ix_data = pd.MultiIndex.from_arrays([ix_event, ix_trial, ix_time])
    
    data_trial = data_trial.set_index(ix_data)
    data_trials.append(data_trial)
data_trials = pd.concat(data_trials)
data_trials = data_trials.sort_index()
```

```python
# Create our data frame
df = data_trials.copy()
df.columns.name = 'name'
use_names = names
df = df[use_names].stack('name').unstack('time')
df = df.replace({np.nan: 0})
```

```python
# Remove trials that are deviant
std_mult = 4
std_cut = df.stack('time').unstack('name').std().max() * std_mult
bad_trials = (df > std_cut).any(1)
df = df[~bad_trials]
```

```python
# First z-score each timeseries
df = df.apply(scale, axis=1)
```

```python
base_stt = -50
base_stp = 0
base_mean = df.loc[:, (df.columns > base_stt) * (df.columns < base_stp)].mean(1)
base_std = df.loc[:, (df.columns > base_stt) * (df.columns < base_stp)].std(1)
```

```python
# Baseline with a z-score
df -= np.tile(base_mean, (df.shape[1], 1)).T
# df /= np.tile(base_std, (df.shape[1], 1)).T
```

```python
# Plot raw timeseries w/o re-aligning
df.mean(level=('event', 'name')).T.plot(figsize=(15, 5), legend=False)
```

```python
# Line up timeseries at min or max values
line_up = 'max'
min_1, min_2 = [200, 400]
use_range = (df.columns > min_1) * (df.columns < min_2)
min_times = pd.DataFrame(index=df.index, columns=['time'])
for ix, vals in df.loc[:, use_range].iterrows():
    if line_up == 'min':
        min_times.loc[ix, :] = vals.argmin()
    elif line_up == 'max':
        min_times.loc[ix, :] = vals.argmax()
min_times -= 300
min_times /= 2
min_times = min_times.squeeze()

df_shift = df.copy()
for ix, shift in min_times.iteritems():
    shifted_values = df.loc[ix, :].shift(-shift)
    df_shift.loc[ix, :] = shifted_values
```

```python
# AUC for individual trials
df_auc = df_shift.query('event!="blink"').loc[:, use_range].sum(1)
```
